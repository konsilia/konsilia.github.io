```python
import time
import numpy as np
import sys
hku_sem1_df_new = pd.DataFrame(columns=sample.columns)
data = {}
for index, row in sem1_df.iterrows():
    if len(row["COURSE CODE"]) > 8:
        continue
    if (index % 250 == 0):
        print("At " + str(index))
    try:
        class_number = int(row["COURSE CODE"][4:8])
        number = row["COURSE CODE"][4:]
        mnemonic = row["COURSE CODE"][:4]
        section = row["CLASS SECTION"]
        courseType = "Course"
        units = 6
        instructor = row["INSTRUCTOR"]
        ###
        days = ["MON","TUE","WED","THU","FRI","SAT","SUN"]
        combinedDays = ""
        for day in days:
            if not pd.isnull(row[day]):
                combinedDays += row[day][:2].title()
        startTime = time.strftime("%I:%M%p", time.strptime(row["START TIME"], "%H:%M"))
        endTime = time.strftime("%I:%M%p", time.strptime(row["END TIME"], "%H:%M"))
        ###
        days1 = '{0} {1} - {2}'.format(combinedDays, startTime, endTime)
        room1 = row["VENUE"]
        ###
        startDate = time.strftime("%m/%d/%Y", time.strptime(row["START DATE"],"%Y-%m-%d"))
        endDate = time.strftime("%m/%d/%Y", time.strptime(row["END DATE"],"%Y-%m-%d"))
        ###
        meetingDates1 = '{0} - {1}'.format(startDate, endDate)
        title = row["COURSE TITLE"]
        description = row["ACAD_CAREER"] + " - " + row["OFFER DEPT"]

        new_row = {
            'ClassNumber': class_number,
            'Days1': "",
            'Days2': "",
            'Days3': "",
            'Days4': "",
            'Description': description,
            'Enrollment': 0,
            'EnrollmentLimit': 100,
            'Instructor1': instructor,
            'Instructor2': "",
            'Instructor3': "",
            'Instructor4': "",
            'MeetingDates1': "",
            'MeetingDates2': "",
            'MeetingDates3': "",
            'MeetingDates4': "",
            'Mnemonic': mnemonic,
            'Number': number,
            'Room1': "",
            'Room2': "",
            'Room3': "",
            'Room4': "",
            'Section': section,
            'Status': 'Open',
            'Title': title,
            'Topic': "",
            'Type': "",
            'Units': '6',
            'Waitlist': 0,
            'times': {}
        }
        key = row["COURSE CODE"]+section
        if not key in data:
            data[key] = new_row
        if not days1 in data[key]['times']:   
            data[key]['times'][days1] = {'Days':days1,'MeetingDates':meetingDates1,'Room':room1}
    except Exception as e:
        continue
for key in data:
    new_row = data[key]
    times = []
    for i in data[key]["times"]:
        times.append(data[key]["times"][i])
    new_row.pop('times', None)
#     if len(times) > 4:
#         print(key)
#         print("There are more times | " + str(len(times)))
    for i in range(min(len(times),4)):
        entry = times[i]
        for attribute in entry:
            new_row[attribute+str(i+1)] = entry[attribute]
    hku_sem1_df_new = hku_sem1_df_new.append(new_row, ignore_index=True)
hku_sem1_df_new_csv = hku_sem1_df_new.replace(np.nan, '', regex=True)
export_csv = hku_sem1_df_new_csv.to_csv (r'../data/Semester Data/CS1198Data.csv', index = None, header=True)
print("Done")
```
